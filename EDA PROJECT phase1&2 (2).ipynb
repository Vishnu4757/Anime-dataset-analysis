# **Checking outliers :**

```python
numeric_cols=df.select_dtypes(include=['number']).columns
numeric_cols
```

```python
print("\n === Outliers Count(IQR method) ===")

for col in numeric_cols:
    Q1= df[col].quantile(0.25)
    Q3= df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col} : {len(outliers)} outliers")
```

### **Observation :*
* Outliers are extreme values that are far away from the rest of the data.
* They are so many outliers in DataSet for some numerical columns

## 1. Categorical vs Categorical

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Type', hue='Rating')
plt.title("Type vs Rating")
plt.xlabel("Type of Anime")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# Observation

* **This Bar chart is used to compare two categorical variables and understand how anime types are distributed across different age ratings.**

* TV-type anime appears most frequently across all age ratings.
* PG-13 is the most common rating category and dominates almost all Type groups.
* Rare rating categories like G and Rx have very low counts regardless of Type.
* Movie, OVA, and ONA formats show smaller but consistent distribution across ratings.

```python
grade_counts = df['Grade'].value_counts()
plt.figure(figsize=(18,5))

# ---------- 1. PIE CHART ----------
plt.subplot(1, 3, 1)
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%')
plt.title("Grade Distribution (Pie Chart)")

# ---------- 2. BAR CHART ----------
plt.subplot(1, 3, 2)
plt.bar(grade_counts.index, grade_counts.values)
plt.title("Grade Count (Bar Chart)")
plt.xlabel("Grade")
plt.ylabel("Count")

# ---------- 3. BOX PLOT ----------
plt.subplot(1, 3, 3)
sns.boxplot(x='Grade', y='Score', data=df,
            order=['Super Hit', 'Hit', 'Average', 'Low'])
plt.title("Score Distribution by Grade (Box Plot)")
plt.xlabel("Grade")
plt.ylabel("Score")

plt.tight_layout()
plt.show()

```

### Observation

* **Pie chart** : percentage of each grade
* Most anime fall under the Low and Average categories, while Super Hit anime form the smallest portion.

* **Bar chart** : frequency of each grade
* The count of Low grade anime is the highest, and Super Hit anime are the fewest, showing that highly popular anime are rare.

* **Boxplot** : how the Score varies across Grade categories
* Shows median, quartiles, and outliers
* Super Hit anime generally have higher scores, while Low grade anime show lower and more spread-out scores.
